```python
import requests
import json
import pandas as pd
```

```python
response = requests.get("http://api.bcb.gov.br/dados/serie/bcdata.sgs.4447/dados?formato=json&dataInicial=01/01/2010&dataFinal=31/12/2016")
```

```python
json = json.dumps(response.json(),sort_keys=True, indent=4)
```

```python
#print(json)
```

```python
with open('json_bc.json', 'w') as arquivo:
    arquivo.write(json)
```
