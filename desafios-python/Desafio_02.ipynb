### Desafio 2

Escreva uma função que recebe um número inteiro de horas e a converte em segundos.

Exemplo:

convert(5) ➞ 18000

convert(3) ➞ 10800

convert(2) ➞ 7200

```python
def convert_to_sec(number):
    minutos = number*60
    segundos = minutos*60
    return segundos

print(convert_to_sec(5))
print(convert_to_sec(3))
print(convert_to_sec(2))
```
