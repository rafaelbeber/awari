### Desafio 10

Dada uma lista, divida-a em 3 partes iguais e reverta cada lista.

Exemplo: 

sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

Parte 1 [8, 45, 11]

Parte 2 [12, 14, 23]

Parte 3 [89, 45, 78]

```python
# Seu c√≥digo

sampleList = [11, 45, 8, 23, 14, 12, 78, 45, 89]

newList = []
for i in range(0, len(sampleList), 3):
  splited = sampleList[i:i+3]
  splited.reverse()
  newList.append(splited)
  
newList
  
```
